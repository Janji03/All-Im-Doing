```python

#Domanda 4

import numpy as np
import scipy.linalg
import matplotlib.pyplot as plt
from skimage import data
from skimage.io import imread

 
A = data.camera()
print(type(A))
print(A.shape)
plt.imshow(A, cmap='gray')
plt.show()


U, s, Vt = scipy.linalg.svd(A)
print('Shape di U:', U.shape)
print('Shape di s:', s.shape)
print('Shape di V:', Vt.shape)


Ap = np.zeros(A.shape)
# numero massimo di diadi
pmax = 20

list_err_rel = []
list_cp = []
for i in range(pmax):
    ui = U[:,i]
    vi = Vt[i,:]
    Ap = Ap + s[i] * np.outer(ui, vi) 
    err_rel = np.linalg.norm(A - Ap) / np.linalg.norm(A)
    list_err_rel.append(err_rel)
    cp = 1 / (i+1) * min(A.shape) - 1
    list_cp.append(cp)
    plt.imshow(Ap, cmap='gray')
    plt.show()



dim_matri_x = np.linspace(0, pmax, pmax)          
                                 
plt.plot(dim_matri_x, list_err_rel)
plt.title('Errore relativo')
plt.xlabel('p')
plt.ylabel('err')
plt.show()

plt.plot(dim_matri_x, list_cp)
plt.title('Fattore di compressione')
plt.xlabel('p')
plt.ylabel('cp')
plt.show()
```
