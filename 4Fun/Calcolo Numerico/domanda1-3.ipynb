```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.linalg
from scipy.linalg import lu_factor 


 ###Domanda 1

n = int(np.random.uniform(10,1000))
A = np.random.randn(n, n)
x = np.ones((n,1))
b = A@x

condA = np.linalg.cond(A, p=np.inf)

print('x: \n', x , '\n')
print('x.shape: ', x.shape, '\n' )
print('b: \n', b , '\n')
print('b.shape: ', b.shape, '\n' )
print('A: \n', A, '\n')
print('A.shape: ', A.shape, '\n' )
print('K(A)=', condA, '\n')

lu, piv =scipy.linalg.lu_factor(A)

print('lu',lu,'\n')
print('piv',piv,'\n')



my_x = scipy.linalg.lu_solve((lu, piv), b)

print('my_x = \n', my_x)
print('norm =', scipy.linalg.norm(x-my_x, 'fro'))


```

```python
#Donanda 2


n = np.random.randint(5, 10)
v = np.full(n, 9) 
A = np.diag(v) + np.diag(np.full(n-1, -4), k=-1) + np.diag(np.full(n-1, -4), k=1)

x = np.ones((n,1))
b = A@x

condA = np.linalg.cond(A)

print('x: \n', x , '\n')
print('x.shape: ', x.shape, '\n' )
print('b: \n', b , '\n')
print('b.shape: ', b.shape, '\n' )
print('A: \n', A, '\n')
print('A.shape: ', A.shape, '\n' )
print('K(A)=', condA, '\n')


L = scipy.linalg.cholesky(A)
print('L:', L, '\n')

print('L.T*L =', L.T@L)
print('err = ', scipy.linalg.norm(A-np.matmul(L, np.transpose(L)), 2))

y = np.linalg.inv(L)@b #Ly = b
my_x = np.linalg.inv(L.T)@y # Ltx = b
print('my_x = \n ', my_x)

print('norm =', scipy.linalg.norm(x-my_x, 'fro'))
```

```python
#Domanda 3 

n = np.random.randint(2, 13)
A = scipy.linalg.hilbert(n)
x = np.ones((n, 1))
b = A@x

condA = np.linalg.cond(A)

print('x: \n', x , '\n')
print('x.shape: ', x.shape, '\n' )
print('b: \n', b , '\n')
print('b.shape: ', b.shape, '\n' )
print('A: \n', A, '\n')
print('A.shape: ', A.shape, '\n' )
print('K(A)=', condA, '\n')

L = scipy.linalg.cholesky(A)
print('L:', L, '\n')

print('L.T*L =', L.T@L)
print('err = ', scipy.linalg.norm(A-np.matmul(L, np.transpose(L)), 2))


y = np.linalg.inv(L)@b
my_x = np.linalg.inv(L.T)@y
print('my_x = \n ', my_x)

print('norm =', scipy.linalg.norm(x-my_x, 'fro'))
```
