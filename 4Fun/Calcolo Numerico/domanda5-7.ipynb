```python
import numpy as np
import scipy.linalg
from scipy.linalg import svd
import matplotlib.pyplot as plt

#Domanda 5

m = 10
x = np.linspace(-1, 1, m)
y = np.exp(x/2)

print('Punti della funzione f(x):')
for i in range(x.size):
    print(f"P{i} = ({x[i]}, {y[i]})")
print('\n')


# grafico funzione esatta ed approssimata
grad = (1, 2, 3, 5, 7)
mplot = 100
xplot = np.linspace(x[0], x[-1], mplot)

for n in grad:
    A = np.zeros((m, n + 1))
    for i in range(n + 1):
      A[:,i] = x**i
    
    # risolvo con SVD
    U, S, V = svd(A)
    alphaSvd = np.zeros(n + 1)
    for i in range(n + 1):
        alphaSvd = alphaSvd + (U[:,i].T@y) * V[i,:] / S[i]
    
    AplotSVD = np.zeros((mplot, n + 1))
    for i in range(n + 1):
        AplotSVD[:,i] = xplot ** i
    yApprSVD = AplotSVD@alphaSvd
    
    
    # risolvo con equazioni normali
    ATA = A.T@A
    ATy = A.T@y
    L, piv = scipy.linalg.lu_factor(ATA)
    alphaEQ = scipy.linalg.lu_solve((L, piv), ATy)
    AplotEQ = np.zeros((mplot, n + 1))
    for i in range(n + 1):
      AplotEQ[:,i] = xplot ** i
    yApprEQ = AplotEQ@alphaEQ
    
    # metto i grafici di SVD e EQ uno di fianco all'altro 
    fig, axs = plt.subplots(1, 2, figsize=(12, 6)) 
    fig.suptitle(f"Confronto Interpolazione polinomiale di grado {n}")

    # Grafico SVD
    axs[0].plot(x, y, 'o')
    axs[0].plot(xplot, yApprSVD, 'r')
    axs[0].set_xlabel('x')
    axs[0].set_ylabel('y')
    axs[0].set_title("Interpolazione polinomiale con SVD")

    # Grafico Equazioni Normali
    axs[1].plot(x, y, 'o')
    axs[1].plot(xplot, yApprEQ, 'r')
    axs[1].set_xlabel('x')
    axs[1].set_ylabel('y')
    axs[1].set_title("Interpolazione polinomiale con Equazioni Normali")

    plt.tight_layout()
    plt.show()
```
