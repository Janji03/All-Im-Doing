```python
import numpy as np
import matplotlib.pyplot as plt

#Domanda 8

# approssimazioni successive
def approssimazioniSuccessive(f, g, tollf, tollx, maxit, xstar, x0 = 0):
  i = 0
  errPassoK = tollx + 1
  errList = np.zeros(maxit + 1, dtype=np.float64)
  errList[0] = np.abs(x0 - xstar)
  xprec = x0

  while (i < maxit and (np.abs(f(xprec)) > tollf or errPassoK > tollx)):
    xk = g(xprec)
    errPassoK = np.abs(xk - xprec)
    i = i + 1
    errList[i] = np.abs(xk - xstar)
    xprec = xk
  errList = errList[0:i]
  return (xprec, i, errList)

# Newton
def newton(f, derivataf, tollf, tollx, maxit, xstar, x0 = 0):
  g = lambda x: x - (f(x) / derivataf(x))
  (x, i, errList) = approssimazioniSuccessive(f, g, tollf, tollx, maxit, xstar, x0)
  return (x, i, errList)

# funzione di cui calcolare lo zero 
f = lambda x: np.exp(x) - x**2
# derivata della funzione
df = lambda x: np.exp(x) - 2*x
# funzione1 utilizzata nel metodo delle approssimazioni successive
g1 = lambda x: x - f(x) * np.exp(x/2)
# funzione2 utilizzata nel metodo delle approssimazioni successive
g2 = lambda x: x - f(x) * np.exp(-x/2)

xstar = -0.703467
fstar = f(xstar)

print("x* =", xstar)
print("f(x*) =", fstar)


# disegnare il grafico di f
xplot = np.linspace(-1, 1)
plt.plot(xplot, f(xplot))
plt.plot(xstar, fstar, 'or', label='Esatta')
plt.grid()
plt.legend()
plt.show()


# calcolare lo zero, con accuratezza e numero di iterazioni
plt.plot(xplot, f(xplot))
plt.plot(xstar, fstar, 'or', label='Esatta')
tollx = 10**(-10)
tollf = 10**(-6)
maxit = 100
x0 = 0

[sol1, iter1, errList1] = approssimazioniSuccessive(f, g1, tollf, tollx, maxit, xstar, x0)
print("\nMetodo delle approssimazioni successive con g1")
print("x =", sol1)
print("Distanza dalla soluzione esatta =", np.abs(sol1 - xstar))
print("Numero di iterazioni =", iter1)

plt.plot(sol1, f(sol1), 'o', label='g1')

[sol2, iter2, errList2] = approssimazioniSuccessive(f, g2, tollf, tollx, maxit, xstar, x0)
print("\nMetodo delle approssimazioni successive con g2")
print("x =", sol2)
print("Distanza dalla soluzione esatta =", np.abs(sol2 - xstar))
print("Numero di iterazioni =", iter2)

plt.plot(sol2, f(sol2), 'og', label='g2')

[solNew, iterNew, errListNew] = newton(f, df, tollf, tollx, maxit, xstar, x0)
print("\nMetodo di Newton")
print("x =", solNew)
print("Distanza dalla soluzione esatta =", np.abs(solNew - xstar))
print("Numero di iterazioni =", iterNew)

plt.plot(solNew, f(solNew), 'ob', label='Newton')

plt.legend()
plt.grid()
plt.show()


# errore vs iterazioni 
plt.plot(errList1, '.-', color='blue')
plt.plot(errList2[:20], '.-', color='green')
plt.plot(errListNew, '.-', color='red')

plt.legend(("g1", "g2", "Newton"))
plt.xlabel('iterazioni')
plt.ylabel('errore')
plt.title("Errore vs iterazioni")
plt.grid()
plt.show()
```
